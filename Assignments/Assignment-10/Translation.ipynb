Translation to 3 Languages (Kannada, Japaneese, French)

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

Load the NLLB-200 distilled model and tokenizer

```python
model = AutoModelForSeq2SeqLM.from_pretrained('facebook/nllb-200-distilled-600M')
tokenizer = AutoTokenizer.from_pretrained('facebook/nllb-200-distilled-600M')
```

Translation function

```python
def translate(input_text, src_lang='eng_Latn', target_lang='kan_Knda'):
    translator = pipeline(
        'translation',
        model=model,
        tokenizer=tokenizer,
        src_lang=src_lang,
        tgt_lang=target_lang,
        max_length=400
    )
    return translator(input_text)[0]['translation_text']
```

Input text

```python
input_text = "My name is Poojitha"
```

Perform translations

```python
kannada = translate(input_text, target_lang='kan_Knda')
japanese = translate(input_text, target_lang='jpn_Jpan')
french = translate(input_text, target_lang='fra_Latn')
```

Display outputs

```python
print(f"Kannada translation: {kannada}\n")
print(f"Japanese translation: {japanese}\n")
print(f"French translation: {french}\n")
```
