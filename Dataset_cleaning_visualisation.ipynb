```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
url = "https://drive.google.com/uc?id=1xQ9SRFer50mFP-y1kI9s4Awj2JWBOEyX"

df = pd.read_csv(url)

# 2. Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# 3. Display basic info
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- First 5 Rows ---")
print(df.head())

# 4. Check for missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())

# 5. Basic statistics
print("\n--- Statistical Summary ---")
print(df.describe())

# 6. Select numerical features for analysis
feature_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

```

```python
# 7. Visualization 1: Pairplot of numerical features
sns.pairplot(df[feature_cols])
plt.suptitle("Pairplot of Numerical Job Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# 8. Visualization 2: Boxplot of Salary by Experience Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_(annual)', data=df, palette='Set3')
plt.title("Annual Salary by Experience Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 9. Visualization 3: Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = df[feature_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Heatmap of Job Features")
plt.tight_layout()
plt.show()
```
